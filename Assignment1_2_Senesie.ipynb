```python

# Import necessary libraries
import pandas as pd  # Pandas for handling and cleaning data
import numpy as np   # NumPy for numerical operations
import matplotlib.pyplot as plt  # Matplotlib for visualization
```

```python

# Load datasets from CSV files
df2 = pd.read_csv("/Users/Mujeh/Downloads/AH_Provisional_Diabetes_Death_Counts_for_2020_20250207.csv")  # Load second dataset
```

```python
# Display the first 5 rows of the dataset to understand its structure
print("First 5 rows of the dataset:")
display(df2.head())  # Display the first few rows
```

```python
# Check the general information about the dataset (column names, data types, missing values, etc.)
print("\nDataset Information:")
df2.info()
```

```python
# Make column names lowercase
# Convert column names to lowercase and replace spaces/special characters
# This is important for SQL compatibility, as SQL does not allow spaces or special characters in column names
df2.columns = df2.columns.str.lower().str.replace(" ", "_").str.replace(r"[^\w]", "", regex=True)
```

```python
# check if code worked
# Display updated column names
print("\nUpdated column names for SQL compatibility:")
print(df2.columns)
```

```python
# Remove all special characters (including underscores) from column names
df2.columns = df2.columns.str.lower().str.replace(r"[^\w]", "", regex=True)

# Display updated column names to verify changes
print("\nUpdated column names after removing all special characters:")
print(df2.columns)
```

```python
# Remove underscores and other special characters from column names
df2.columns = (df2.columns
    .str.lower() # Convert to lowercase
    .str.replace(r"[^a-zA-Z0-9]", "", regex=True)  # Remove all non-alphanumeric characters
              )
# Display updated column names to verify changes
print("\nUpdated column names after removing underscores and special characters:")
print(df2.columns)
```

```python
# Check for missing values in each column
print("\nMissing Values Before Cleaning:")
print(df2.isnull().sum())
```

```python
# Handle missing values
# Strategy: If a column has more than 50% missing values, drop it. Otherwise, fill missing values appropriately.

threshold = len(df2) * 0.5  # Define a threshold (50% of the dataset size)
df2 = df2.dropna(thresh=threshold, axis=1)  # Drop columns with more than 50% missing values
```

```python
# Fill missing values with the median for each numeric column safely
for col in df2.select_dtypes(include=['number']).columns:  # Select only numeric columns
    df2[col] = df2[col].fillna(df2[col].median())  # Assign back explicitly
```

```python
# Check if there are any date columns and convert them to datetime format
date_cols = [col for col in df2.columns if "date" in col.lower()]  # Identify columns with 'date' in their name
for col in date_cols:
    df2[col] = pd.to_datetime(df2[col], errors="coerce")  # Convert to datetime format, ignore errors
```

```python
# Verify changes in data types
print("\nData Types After Cleaning:")
print(df2.dtypes)
```

```python
# Ensure df2 is an independent copy (if needed)
df2 = df2.copy()  

# Fill missing values in numerical columns with the median (corrected)
for col in df2.select_dtypes(include=['number']).columns:
    df2.loc[:, col] = df2[col].fillna(df2[col].median())  # Safe assignment

# Fill missing values in categorical columns with 'Unknown' (corrected)
for col in df2.select_dtypes(include=['object']).columns:
    df2.loc[:, col] = df2[col].fillna("Unknown")  # Safe assignment without inplace=True

# Verify no missing values remain
print("\nMissing Values After Cleaning:")
print(df2.isnull().sum())
```

```python
# Remove duplicate rows (if any exist) to avoid redundant data
df2.drop_duplicates(inplace=True)
```

```python
 # Standardize categorical data: Convert all text to lowercase to ensure consistency
for col in df2.select_dtypes(include=['object']).columns:  # Loop through all categorical columns
    df2[col] = df2[col].str.lower().str.strip()  # Convert text to lowercase and remove leading/trailing spaces
```

```python
# Check if there are any date columns and convert them to datetime format
date_cols = [col for col in df2.columns if "date" in col.lower()]  # Identify columns with 'date' in their name
for col in date_cols:
    df2[col] = pd.to_datetime(df2[col], errors="coerce")  # Convert to datetime format, ignore errors
```

```python
# Verify changes in data types
print("\nData Types After Cleaning:")
print(df2.dtypes)
```

```python
# Save the cleaned dataset to a new CSV file
cleaned_file_path = "/Users/Mujeh/Downloads/cleaned_diabetes_data.csv"  # Define save location
df2.to_csv(cleaned_file_path, index=False)  # Save the cleaned data without row indices

print(f"\nDataset cleaned and saved as: {cleaned_file_path}")  # Confirm successful save
```

```python
# Basic visualization to check categorical data distribution
for col in df2.select_dtypes(include=['object']).columns:  # Loop through categorical columns
    if df2[col].nunique() < 15:  # Only plot if the column has fewer than 15 unique categories
        plt.figure(figsize=(10, 5))  # Set figure size
        df2[col].value_counts().plot(kind="bar", color="skyblue", edgecolor="black")  # Create a bar chart
        plt.title(f"Distribution of {col}")  # Add title
        plt.xlabel(col)  # Label x-axis
        plt.ylabel("Count")  # Label y-axis
        plt.xticks(rotation=45)  # Rotate x-axis labels for readability
        plt.show()  # Display the plot
```
